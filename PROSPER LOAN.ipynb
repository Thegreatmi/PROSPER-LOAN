##  QUESTION 3: DOES THE INTEREST RATE HAS ANY CORRELATION WITH THE MONTHLY LOAN PAYMENT (2 QUANTITATIVE VARIABLE).

```python
sns.regplot(data = df, x='MonthlyLoanPayment', y='BorrowerRate', fit_reg = 'false');
plt.title('BORROWERRATE EFFECT ON LOAN PAYMENT')
```

OBSERVATION: The scattered plot is used to show correlation between two variables which in this case they are interest rate and Monthly loan repayment. this scatter plot shows a downward movement of the regression line , this simply means that as the loan collected or recover by Prosper Loan company increases over the month , theres also a downward review of interest rate , this basically indicate that theres a positive correlation between the two variables. i.e 'BORROWER RATE AND MONTHLY LOAN REPAYMENT".
